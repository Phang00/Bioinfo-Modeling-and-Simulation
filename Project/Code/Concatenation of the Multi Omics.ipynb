```python
# Import necessary libraries
import os
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/gdrive')

# Set the data path for dataset and model location
dataset_dir = "/content/gdrive/My Drive/PROJECT/"
model_loc = "/content/gdrive/My Drive/PROJECT/"

# Check the contents of the dataset directory
print(os.listdir(dataset_dir))

# Load the data from CSV files
mrna_data = pd.read_csv(dataset_dir + 'BRCA_mRNA.csv', index_col=0)
miRNA_data = pd.read_csv(dataset_dir + 'BRCA_miRNA.csv', index_col=0)
DNA_Methylation_data = pd.read_csv(dataset_dir + 'BRCA_DNA_Methylation.csv', index_col=0)
copy_number_variation_data = pd.read_csv(dataset_dir + 'BRCA_Copy Number Variation.csv', index_col=0)

# Concatenate the data along the columns (axis=1)
combined_data = pd.concat([mrna_data, miRNA_data, DNA_Methylation_data, copy_number_variation_data], axis=1)

# Save the combined data to a CSV file
combined_data.to_csv(dataset_dir + 'BRCA_Multi_Omics.csv')

# Print a message indicating that the file has been saved
print("Combined data has been saved to BRCA_Multi_Omics.csv")
```
